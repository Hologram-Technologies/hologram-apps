# Holo Geometric — geometric algebra in your browser (kingdon)

Completes the novel-computation set: **kingdon** (geometric / Clifford algebra) runs offline in Pyodide. Rotations, reflections, and bivectors as algebra — no trig bookkeeping. Run all cells.

```python
import piplite
await piplite.install(['kingdon'])
from kingdon import Algebra
import numpy as np, math
print('kingdon ready (offline)')
```

## 1 · Rotate a vector with a rotor (2D)
A rotor `R = cos(θ/2) − sin(θ/2)·e₁₂` rotates by the sandwich `R v R⁻¹`.

```python
alg = Algebra(2)
e1, e2, e12 = alg.blades.e1, alg.blades.e2, alg.blades.e12
v = 3*e1 + 1*e2
th = math.pi/2
R = math.cos(th/2) - math.sin(th/2)*e12
v_rot = R * v * ~R
print('v        =', v)
print('rotated  =', v_rot, '  (90 deg: (3,1) -> (-1,3))')
```

## 2 · Reflect a vector (sandwich with a unit vector)

```python
n = e1                      # reflect across the e1 axis
v_ref = n * v * n
print('reflected=', v_ref, '  ((3,1) -> (3,-1))')
```

## 3 · The bivector — oriented area from the geometric product

```python
a = 2*e1 + 0*e2
b = 1*e1 + 3*e2
print('a*b      =', a*b, '  (scalar dot + e12 area)')
print('area     =', (a*b).grade(2), '  (signed parallelogram area as a bivector)')
```

### Why this matters here
- Geometric algebra unifies vectors, rotations, and areas in one product — natural for the non–von-Neumann experiments this app is for.
